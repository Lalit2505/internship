```python
rank=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
for i in rank:
    Rank=i.text
    Rank_title.append(Rank)

name=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
for i in name:
    Name=i.text
    Name_title.append(Name)
    
artist=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[3]')
for i in artist:
    Artist=i.text
    Artist_title.append(Artist) 
    
upload=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
for i in name:
    Upload=i.text
    Upload_date.append(Upload) 

views=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
for i in views:
    Views=i.text
    Views_title.append(Views)     
```

```python
print(len(Rank_title),len(Name_title),len(Artist_title),len(Upload_date),len(Views_title))
```
